# 06 kNN Handjob

```python
# 加载数据集
import numpy as np
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data
y = iris.target
target_dict = {0:'setosa',1:'versicolor',2:'virginica'}
```

## Train-Test Split

### 常规思路

先将特征矩阵X与目标向量y合并

进行Shuffle

最后再进行拆分

```python
# 让y与X维度相同，也变成二维
y = y.reshape(X.shape[0], -1)
# 合并X与y,此时y中元素同X，变成了float型
concentrate = np.concatenate([X, y], axis=1)
# 对合并后矩阵进行乱序
np.random.shuffle(concentrate)
# 对数据集进行拆分,注意Y_train和Y_test需要再转化成一维的，并转化为int型
X_train = concentrate[:100,:4]
Y_train = concentrate[:100,4:][:,0].astype(int)
X_test = concentrate[100:,4:]
Y_test = concentrate[100:,4:][:,0].astype(int)
```

### 更好做法

利用permutation生成乱序索引，范围从0到样本个数-1

利用numpy FancyIndexing的特性使用乱序索引访问原数据集

自然得到乱序后得数据集，无需合并

```python
shuffle_indexs = np.random.permutation(y.size)
train_ratio = 0.8
train_num = int(train_ratio*y.size)
y[shuffle_indexs]
X_train = X[shuffle_indexs][:train_num]
Y_train = y[shuffle_indexs][:train_num]
X_test = X[shuffle_indexs][train_num:]
Y_test = y[shuffle_indexs][train_num:]
```

### 调用sklearn

```python
from sklearn.model_selection import train_test_split
# test_size指定测试数据集占比，默认为0.2
# random_state用于指定random过程的seed
# 指定random_state后可保证每次随机过程一致，均是在random_state的基础上完成的
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.2)
X_test
```

## kNN分步实现

```python
# 返回一个测试点x_test与所有样本点的距离，List
def cal_distance(X_train, x_test):
    from math import sqrt
    distances = []
    for x in X_train:
        # 计算欧拉距离
        dis = sqrt(np.sum((x_test-x)**2))
        distances.append(dis)
    return distances
```

```python
# 指定一个k，求出x_distances中距离最近的k个预测分类
# 对这些分类进行投票，返回票数最多的分类,predic_classify是target_dict的key

def kNN_classify(k,x_distances,Y_train,target_dict):
    '''
    target_dict:分类代码与分类的对应关系字典
    x_distances:测试点与测试集所有点的距离列表
    Y_train:训练集的分类结果
    '''
    from collections import Counter
    
    # 对距离列表进行排序，返回元素所在位置
    x_index = np.argsort(x_distances)
    
    # 在训练集中取出前k个距离最近的样本点的索引下标
    topk_X = x_index[:k]
    
    # 得到这k个样本点的分类
    classification_k = Y_train[topk_X]
    
    # 对k个分类结果进行投票，最终结果票多者胜
    # 各个分类结果的票数:字典
    votes_dict = Counter(classification_k)
    
    # Counter的most_common(n)函数用于找出票数最高的n的元素,返回为List，List中的每个元素为tuple
    # 取出票最高者
    predic_classify = votes_dict.most_common(1)[0][0]
    return target_dict[predic_classify]
```

```python
# 选择一个测试点
x_test = X_test[:1,:]

# 计算测试点与训练集的距离列表
x_distances = cal_distance(X_train, x_test)

# 指定k=6得到预测分类结果
k=6
kNN_classify(k, x_distances, Y_train, target_dict)
```

## 五行封装

```python
# 使用生成式 五行核心代码搞定knn
def kNN(k,X_train,Y_train,x_test,target_dict):
    from math import sqrt
    import numpy as np
    from collections import Counter
    # line1：测试点x_test的距离列表
    x_distances = [sqrt( np.sum( (x_test-x_train)**2 ) ) for x_train in X_train]
    # line2: 排序后得索引取前k，得其分类结果
    classification_k = Y_train[np.argsort(x_distances)[:k]]
    # line3：进行分类结果投票
    votes_dict = Counter(classification_k)
    # line4：票高者胜
    predic_classify = votes_dict.most_common(1)[0][0]
    # line5：返回具体类型
    return target_dict[predic_classify]
```

```python
# 测试
x_test = X_test[:2,:]
k = 6
kNN(k,X_train,Y_train,x_test,target_dict)
```

## 调用sklearn

```python
from sklearn.neighbors import KNeighborsClassifier
# kNN分类器初始化
knn_classifier = KNeighborsClassifier(n_neighbors=6)
# 模型训练
knn_classifier.fit(X_train, Y_train)
# 模型预测
y_predict = knn_classifier.predict(X_test)
```

```python
# 测试数据集的准确率
accuracy = sum(y_predict==Y_test)/Y_test.size
accuracy
```

## kNN超参数确定


```python
# 实验验证k
best_k = -1
best_score = 0.0
for k in range(1,11):
    knn_classifier = KNeighborsClassifier(n_neighbors=k)
    knn_classifier.fit(X_train,Y_train)
    score = knn_classifier.score(X_test, Y_test)
    if score>best_score:
        best_score = score
        best_k = k
print('best_k :', best_k )
print('best_score: ', best_score)
```

```python
# kNN的另一个超参数 距离权重weights
# weights默认为uniform，不设置权重
# weights = distance ,将距离倒数设为票数的权重，有效解决平票问题
best_method = ''
best_k = -1
best_score = 0.0
for method in ['uniform', 'distance']:
    for k in range(1,11):
        knn_classifier = KNeighborsClassifier(n_neighbors=k, weights=method)
        knn_classifier.fit(X_train,Y_train)
        score = knn_classifier.score(X_test, Y_test)
        if score>best_score:
            best_score = score
            best_k = k
            best_method = method
print('best_k :', best_k )
print('best_score: ', best_score)
print('best_method: ', best_method)
```

```python
%%time
# kNN的另一个超参数，明科夫斯基距离p
# Σ|Xa - Xb|{p * 1/p}
# p = 1 曼哈顿距离（一维）
# p = 2 欧拉距离 （二维）
# p... 高维
# 这是一个k*p的网格搜索，此时weights需取distance设置p才有效
best_p = -1
best_k = -1
best_score = 0.0
for k in range(1,11):
    for p in range(1,5):
        knn_classifier = KNeighborsClassifier(n_neighbors=k, weights='distance', p=p)
        knn_classifier.fit(X_train,Y_train)
        score = knn_classifier.score(X_test, Y_test)
        if score>best_score:
            best_score = score
            best_k = k
            best_p = p
print('best_k :', best_k )
print('best_p :', best_p )
print('best_score: ', best_score)
```

## kNN网格搜索确定超参

```python
# 为网格搜索确定参数列表，列表中每个元素是一个字典，代表一组搜索
grid_parms = [
    {
        'weights':['uniform'],
        'n_neighbors': [i for i in range(1,11)]
    },
    {
        'weights':['distance'],
        'n_neighbors': [i for i in range(1,11)],
        'p': [i for i in range(1,11)]
    }
]
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
knn_clf = KNeighborsClassifier()
```

```python
%%time
grid_search = GridSearchCV(knn_clf, grid_parms)
grid_search.fit(X_train, Y_train)
# n_jobs默认为1 单线程
# n_jobs = -1 当前计算机cpu所有内核全部进行运算
```

```python
%%time
# verbose训练过程中显示输出提示，输入的整数值越大，返回信息越多
grid_search = GridSearchCV(knn_clf, grid_parms,n_jobs=-1, verbose=2 )
grid_search.fit(X_train, Y_train)
```

```python
# 返回搜索到的最佳模型
grid_search.best_estimator_
```

```python
# 返回测试准确率，测试方法为交叉验证CV
grid_search.best_score_
```

```python
# 返回搜索到的最佳超参数
grid_search.best_params_
```

```python
# 使用最佳模型进行预测
knn_clf = grid_search.best_estimator_
y_predict = knn_clf.predict(X_test)
y_target = [target_dict[y] for y in y_predict]
y_target
```

```python
# 最佳模型在实际预测过程中的准确率
knn_clf.score(X_test,Y_test)
```
